```python
import os
import random
import pickle
import cv2
import pandas as pd
import numpy as np
import keras
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
```

```python
working_directory = os.getcwd()
```

```python
data_set = os.path.join(working_directory, 'Dataset')
```

```python
os.listdir(data_set)
```

```python
csv = pd.read_csv(os.path.join(data_set, 'HAM10000_metadata.csv'))
```

```python
csv.head()
```

```python
part1 = os.path.join(data_set, 'Part1')
```

```python
def data_generator(width, height):
    data = []
    try:
        for index, row in csv.iterrows():
            y = row['dx']
            print(y)
            image_id = row['image_id']
            #print(image_id)
            image_location = os.path.join(part1, image_id + ".jpg")
            #print(image_location)
            img_array = cv2.imread(image_location, -1)
            #plt.imshow(img_array)
            new_array = cv2.resize(img_array, (height, width))
            data.append([new_array, y])
    except Exception as e:
        pass
    return data
```

```python
training_data = data_generator(50,50)
```

```python
len(training_data)
```

```python
random.shuffle(training_data)
```

```python
def feature_extractor(data, X, y):
    for feature, label in data:
        X.append(feature)
        y.append(label)
```

```python
X_train, y_train = [], []
feature_extractor(training_data, X_train, y_train)
```

```python
X = np.array(X_train).reshape(len(training_data), 50, 50, 3)
```

```python
encoder = LabelEncoder()
encoded_y = encoder.fit_transform(y_train)
one_hot_y_train = to_categorical(encoded_y)
```

```python
pickle_out = open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(one_hot_y_train, pickle_out)
pickle_out.close()
```
